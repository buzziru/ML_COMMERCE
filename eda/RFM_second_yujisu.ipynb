- 최종결제금액 컬럼 생성
  - 미국은 세금이 물건 가격에 포함되어 있지 않으므로 판매세를 구하여 따로 더함

```python
# 금액에 쿠폰 적용 상태와 할인율, 세금을 적용하여 최종 결제 금액을 계산

onlinesales_merge['할인율'] = onlinesales_merge.apply(lambda row: 0 if pd.isna(row['할인율']) else row['할인율'], axis=1)
onlinesales_merge['할인된_금액'] = onlinesales_merge.apply(lambda row: row['평균금액'] * (1 - row['할인율'] / 100) if row['쿠폰상태'] == 'Used' and pd.notna(row['쿠폰코드']) else row['평균금액'], axis=1)
onlinesales_merge['판매세'] = onlinesales_merge['할인된_금액'] * onlinesales_merge['GST']
onlinesales_merge['최종결제금액'] = (onlinesales_merge['할인된_금액'] + onlinesales_merge['판매세']) * onlinesales_merge['수량']
```

```python
# 거래ID당 한번만 배송료 계산
def deliveryCal(df):

  unique_df = df.drop_duplicates(['고객ID', '거래ID'])

  unique_df['최종결제금액'] = unique_df['최종결제금액'] + unique_df['배송료']

  df_combined = pd.concat([unique_df, df[df.duplicated(subset=['고객ID', '거래ID'], keep='first')]])

  return df_combined

onlinesales_merge = deliveryCal(onlinesales_merge)
onlinesales_merge = onlinesales_merge.drop(['할인된_금액','판매세'], axis=1)
```

# 4. RFM 분석

- R : 얼마나 최근에 방문했는가
- F : 얼마나 자주 방문했는가
- M : 얼마나 많은 돈을 사용했는가

  - LRFMV 모델
    - L : 고객의 첫번째 방문과 마지막 방문 사이의 일수
    - V : 고객이 하루 동안 구매한 평균 상품 수

## 4-1. RFM 스코어 열 생성을 위한 작업

```python
onlinesales_merge.groupby('고객ID')['거래날짜'].max().reset_index()
```

```python
rfm_online_sales = onlinesales_merge.copy()

recency_data = rfm_online_sales.groupby('고객ID')['거래날짜'].max().reset_index()
recency_data['Recency'] = (rfm_online_sales['거래날짜'].max() - recency_data['거래날짜']).dt.days + 1 # +1 : 최근 거래 날짜가 2019-12-31일 때 Recency값이 0이 나오는 경우를 처리하기 위해

frequency_data = rfm_online_sales.groupby('고객ID')['거래ID'].count().reset_index()
frequency_data.rename(columns={'거래ID': 'Frequency'}, inplace=True)

monetary_data = rfm_online_sales.groupby('고객ID')['최종결제금액'].sum().reset_index()
monetary_data.rename(columns={'최종결제금액': 'Monetary'}, inplace=True)

rfm_data = recency_data.merge(frequency_data, on='고객ID').merge(monetary_data, on='고객ID')
rfm_data.head()
```

- 각 요소별 분포 시각화

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (15, 5))

plt.subplot(1, 3, 1)
sns.histplot(rfm_data['Recency'], bins=20, color='coral', edgecolor='black')
plt.title('Recency (R) Distribution')

plt.subplot(1, 3, 2)
sns.histplot(rfm_data['Frequency'], bins=20, color='skyblue', edgecolor='black')
plt.title('Frequency (F) Distribution')

plt.subplot(1, 3, 3)
sns.histplot(rfm_data['Monetary'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Monetary (M) Distribution')

plt.tight_layout()
plt.show()
```

- --> F와 M은 매우 편향된 경향을 보이는 반면 R은 비교적 고르게 분포하고 있음

- F, M은 왜도와 첨도가 심한 관계로 로그 변환

```python
import numpy as np
```

```python
# F, M 에 대해 로그변환한 후 새 열에 추가하기
# '거래ID_고유개수' 열에 대해 로그 변환 수행
rfm_data['Frequency_log'] = np.log(rfm_data['Frequency'])

# '고객단순구입액' 열에 대해 로그 변환 수행
rfm_data['Monetary_log'] = np.log(rfm_data['Monetary'])
```

- --> 하지만 RFM 스코어 계산시 로그 변환이 큰 의미는 없을 것 같아 로그변환 미실시

```python
rfm_data.drop(['Frequency_log', 'Monetary_log'], axis=1, inplace=True)
```

- R, F, M 열 생성(5점 척도)
  - https://documentation.bloomreach.com/engagement/docs/rfm-segmentation
  

```python
# Recency, Frequency, Monetary를 5분위로 나누어 각각에 점수 부여하는 함수 정의 -> 1~5점 부여

def assign_score(series):
    quartiles = pd.qcut(series,
                        q=5,
                        labels=False, # 각 분위를 숫자로 표시
                        duplicates='drop') # 중복 값이 있는 경우 중복 값을 삭제

    # Recency에 대한 점수는 작은 값일수록 높은 점수를 가지도록 조정
    if series.name == 'Recency':
        return 5 - quartiles
    else:
        return quartiles + 1

# RFM 점수 부여
rfm_data['R'] = assign_score(rfm_data['Recency'])
rfm_data['M'] = assign_score(rfm_data['Monetary'])
rfm_data['F'] = assign_score(rfm_data['Frequency'])
```

```python
rfm_data.drop(['M_log', 'F_log'], axis=1, inplace=True)
```

```python
rfm_data
```

- RFM 스코어 열 생성

```python
# rfm_score을 합치기 위해 string으로 형 변환
rfm_data['R'] = rfm_data['R'].astype(str)
rfm_data['F'] = rfm_data['F'].astype(str)
rfm_data['M'] = rfm_data['M'].astype(str)

# rfm_score 합치기
rfm_data['RFM'] = rfm_data['R'] + rfm_data['F'] + rfm_data['M']
rfm_data
```

## 4-2. RFM 기반 고객 세분화

- VIP고객, 관심 필요 고객, 이탈 고객 등 RFM 점수에 따라 11가지의 고객 유형 기준을 정한 후, 유저들을 세분화
  - 고객의 기준은 해당 링크를 참고하여 세그먼트
  - https://documentation.bloomreach.com/engagement/docs/rfm-segmentation
  
    - 고객 유형의 특징은 다음과 같다
      - VIP 고객: 최근에 구매하고, 자주 주문하고, 가장 많이 소비한 고객
      - 충성 고객: 주기적으로 구매하고, 프로모션에 반응을 잘하는 고객
      - 잠재 충성 고객: 많은 금액의 물품을 구매한 최근 고객

    - 신규 고객: 최근 구매 -> R^ F- M-
    - 관심 필요 고객: 예전에 많이 자주 구매 -> R↓ F^ M^
    - 이탈 우려 고객: 적게 구매 or 자주 구매 X -> R- F- M↓ or R↓ F↓ M^
    - 이탈 고객: 아주 예전에 적개 구매 -> R↓ F↓ M↓

- 고객 세그먼트 함수 정의

```python
# 고객 세그먼트를 결정하는 함수 정의
def classify_customer_segment(row):

    # RFM 값을 문자열로 가져옴
    rfm_value = str(row['RFM'])

    # RFM 값에 따라 고객 세그먼트
    if rfm_value in ['111', '112', '121', '131', '141', '151']:
        return '이탈 고객'

    elif rfm_value in ['255', '254', '245', '244', '253', '252', '243', '242', '235', '234', '225', '224', '153', '152', '145', '143', '142', '135', '134', '133', '125', '124', '332', '322', '233', '232', '223', '222', '132', '123', '122', '212', '211']:
        return '이탈 우려 고객'

    elif rfm_value in ['535', '534', '443', '434', '343', '334', '325', '324', '525', '524', '523', '515', '514', '513', '425', '424', '413', '414', '415', '315', '314', '313', '331', '321', '312', '221', '213', '231', '241', '251', '155', '154', '144', '214', '215', '115', '114', '113']:
        return '관심 필요 고객'

    elif rfm_value in ['553', '551', '552', '541', '542', '533', '532', '531', '452', '451', '442', '441', '431', '453', '433', '432', '423', '353', '352', '351', '342', '341', '333', '323', '522', '521', '512', '511', '422', '421', '412', '411', '311']:
        return '신규 고객'

    elif rfm_value in ['543', '444', '435', '355', '354', '345', '344', '335']:
        return '충성 고객'

    elif rfm_value in ['555', '554', '544', '545', '454', '455', '445']:
        return 'VIP 고객'

    else:
        return '기타'

# 고객 세그먼트를 결정
rfm_data['고객 세그먼트'] = rfm_data.apply(classify_customer_segment, axis=1)
rfm_data
```

- 고객 세그먼트별 유저 수

```python
result_segment_df = rfm_data.groupby('고객 세그먼트')['고객ID'].count().reset_index().rename(columns={'고객ID' : '고객수'})
result_segment_df
```
